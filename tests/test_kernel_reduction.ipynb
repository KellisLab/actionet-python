```python
import actionet
import scanpy as sc
import anndata
import pandas as pd
import numpy as np

import lets_plot
lets_plot.LetsPlot.setup_html(no_js=True)
```

```python
adata = anndata.read_h5ad("../data/test_adata.h5ad")
```

```python
actionet.filter_anndata(adata, min_cells_per_feat=0.01, inplace=True)
sc.pp.normalize_total(adata, target_sum=1e4, inplace=True)
sc.pp.log1p(adata, base=2, copy=False)
```

```python
actionet.reduce_kernel(adata, n_components=30, key_added='action', inplace=True, svd_algorithm="halko")
```
